```python
import numpy as np
import matplotlib.pyplot as plt
```

# Columna A

```python
# Leer los datos del archivo
raster1_colA = np.loadtxt(r'F:\neuroscience_project\Fig_5_Col_A\spikes_multipop1.dat')
raster2_colA = np.loadtxt(r'F:\neuroscience_project\Fig_5_Col_A\spikes_multipop2.dat')
AVG_data_colA = np.loadtxt(r'F:\neuroscience_project\Fig_5_Col_A\averages_multipop1.dat')

tiempos1_A = raster1_colA[:, 0] 
neuronas1_A = raster1_colA[:, 1] 
tiempos2_A = raster2_colA[:, 0] 
neuronas2_A = raster2_colA[:, 1] 

X_1_A = AVG_data_colA[:, 0] 
U_1_A = AVG_data_colA[:, 1] 
X_2_A = AVG_data_colA[:, 2] 
U_2_A = AVG_data_colA[:, 3] 
t_A = AVG_data_colA[:, 4] 
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(2, 1, 1)
plt.plot(t_A,X_1_A, linewidth=7, alpha= 0.2)
plt.plot(t_A,X_2_A, linewidth=7, alpha= 0.2)
plt.xlim(-500,)

plt.subplot(2, 1, 2)
plt.plot(t_A,U_1_A, linewidth=7, alpha= 0.2)
plt.plot(t_A,U_2_A, linewidth=7, alpha= 0.2)
plt.xlim(-500,)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.subplot(2, 1, 1)
plt.plot(tiempos1_A, neuronas1_A, 'ko',markersize=1.5, color='steelblue')
plt.ylim(1600,2000)
plt.xlim(-500,)
plt.subplot(2, 1, 2)
plt.plot(tiempos2_A, neuronas2_A, 'ko',markersize=1.5, color='orange')
plt.ylim(1600,2000)
plt.xlim(-500,)
plt.tight_layout()
plt.show()
```
